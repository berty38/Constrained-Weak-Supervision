```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
n = 100
d = 2
y = np.random.rand(n)
x = np.random.randn(d, n)
```

```python
resolution = 100
values = np.linspace(-10, 10, resolution)
obj = np.zeros((resolution, resolution))

for i in range(resolution):
    for j in range(resolution):
        
        w = np.array([values[i], values[j]])
        
        p = 1.0 / (1 + np.exp(- w.T.dot(x)))
        
        score = y.dot(1 - p) + p.dot(1 - y)
        
        obj[i, j] = score        
```

```python
X, Y = np.meshgrid(values, values)
CS = plt.contour(X, Y, obj, levels=np.linspace(np.min(obj), np.max(obj), 20))
plt.colorbar();
```

```python
contours = plt.contour(X, Y, obj, levels=np.linspace(np.min(obj), np.max(obj), 20))
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(obj, extent=[-10, 10, -10, 10], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, obj, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```
